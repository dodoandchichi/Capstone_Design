```python
import pandas as pd
import numpy as np
from keras.models import Sequential, Model
from keras.layers import LSTM, Dense, Dropout, Input, Permute, Reshape, Multiply, Lambda, Concatenate
from keras import backend as K
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import tensorflow as tf

# Attention  
def attention_layer(inputs, neurons):
    x = Permute((2, 1))(inputs)
    x = Dense(neurons, activation="softmax")(x)
    x = Permute((2, 1), name='attention_weights')(x)
    output = Multiply()([inputs, x])
    return output

#   
normal_data = pd.read_csv("./csv/5000hz/300rpm/300rpm normal data/stream2024_4_22_23_19.csv")
carriage_data = pd.read_csv("./csv/5000hz/300rpm/300rpm high-speed damage/stream2024_4_23_0_20.csv")

#    0() ,    1() 
normal_data['label'] = 0
carriage_data['label'] = 1

#  
combined_data = pd.concat([normal_data, carriage_data], ignore_index=True)

# X, Y, Z    
features = combined_data[['motor1_x', 'motor1_y', 'motor1_z', 'sound']]
labels = combined_data['label']

#  
scaler = StandardScaler()
X = scaler.fit_transform(features)

# /  
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=30)

# LSTM     
# ( ,  ,  )
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

#  
inputs = Input(shape=(X_train.shape[1], X_train.shape[2]))
lstm_out = LSTM(64, return_sequences=True)(inputs)
attention_out = attention_layer(lstm_out, X_train.shape[1])

# Lambda    , output_shape  .
def sum_over_time(x):
    return tf.reduce_sum(x, axis=1)

attention_out = Lambda(sum_over_time, output_shape=(64,))(attention_out)
output = Dense(1, activation='sigmoid')(attention_out)
model = Model(inputs=[inputs], outputs=[output])

#  
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Early stopping 
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

#  
history = model.fit(X_train, y_train, epochs=100, batch_size=128, validation_data=(X_test, y_test), callbacks=[early_stopping])

#  
model.summary()

#     
train_loss = history.history['loss']
val_loss = history.history['val_loss']
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

#       
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(train_loss, label='Train Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

#       
plt.subplot(1, 2, 2)
plt.plot(train_accuracy, label='Train Accuracy')
plt.plot(val_accuracy, label='Validation Accuracy')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()

#  
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_accuracy}')
```

```python
import pandas as pd
import numpy as np
from keras.models import Model
from keras.layers import LSTM, Dense, Dropout, Input, Multiply, Permute, Lambda
from keras import backend as K
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
import tensorflow as tf
import matplotlib.pyplot as plt

# GPU 
physical_devices = tf.config.list_physical_devices('GPU')
if len(physical_devices) == 0:
    print("GPU   . CPU  .")
else:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)

# Attention  
def attention_layer(inputs, neurons):
    x = Permute((2, 1))(inputs)
    x = Dense(neurons, activation="softmax")(x)
    x = Permute((2, 1))(x)
    output = Multiply()([inputs, x])
    return output

#      
normal_data = pd.read_csv("./csv/5000hz/300rpm/300rpm normal data/stream2024_4_22_23_19.csv")
normal_data['label'] = 'normal'

carriage_data = pd.read_csv("./csv/5000hz/300rpm/300rpm carriage damage/stream2024_4_23_2_54.csv")
carriage_data['label'] = 'carriage_data'

highspeed_data = pd.read_csv("./csv/5000hz/300rpm/300rpm high-speed damage/stream2024_4_23_0_20.csv")
highspeed_data['label'] = 'highspeed_data'

corrosin = pd.read_csv("./csv/5000hz/300rpm/300rpm oxidation and corrosion/stream2024_4_23_1_16.csv")
corrosin['label'] = 'corrosin'

# "lack of lubrication"   .

#  
combined_data = pd.concat([normal_data, carriage_data, highspeed_data, corrosin], ignore_index=True)

# X, Y, Z    
features = combined_data[['motor1_x', 'motor1_y', 'motor1_z', 'sound']]
labels = combined_data['label']

#  
scaler = StandardScaler()
X = scaler.fit_transform(features)

#    - 
encoder = LabelEncoder()
y = encoder.fit_transform(labels)
y = to_categorical(y)

# /  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)

# LSTM     
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

#  
inputs = Input(shape=(X_train.shape[1], X_train.shape[2]))
lstm_out = LSTM(64, return_sequences=True)(inputs)
attention_out = attention_layer(lstm_out, 1)  # Attention layer 

# Attention     
def sum_over_time(x):
    return tf.reduce_sum(x, axis=1)

attention_out = Lambda(sum_over_time, output_shape=(64,))(attention_out)
output = Dense(y_train.shape[1], activation='softmax')(attention_out)
model = Model(inputs=[inputs], outputs=[output])
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Early stopping 
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

#  
history = model.fit(X_train, y_train, epochs=100, batch_size=256, validation_data=(X_test, y_test), callbacks=[early_stopping])

#  
model.summary()

#        
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(train_loss, label='Train Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_accuracy, label='Train Accuracy')
plt.plot(val_accuracy, label='Validation Accuracy')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()

#  
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_accuracy}')
```
